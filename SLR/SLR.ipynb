```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
def LinearRegression(x,y):
    meanX = np.mean(x);
    meanY = np.mean(y);
    
    num=0
    den=0
    for i in range(len(x)):
        num+=(x[i]-meanX)*(y[i]-meanY)
        den+=(x[i]-meanX)*(x[i]-meanX)
    slope = num/den
    intercept = meanY - meanX*slope
    
    return (slope,intercept)
```

```python
def PlotGraph(x,y,slope,intercept):
    print('the intercept is :',intercept)
    print('the slope is :',slope)
    
    plt.scatter(x, y, color = "m", marker = "o", s = 30)
    
    plt.plot(x, x*slope + intercept, color = "g") 
    
    plt.xlabel("size (ft^2)") 
    plt.ylabel("Rental-Cost ($)") 
    
    plt.show()
```

```python
X = []
Y = []
m = 0
q = 1

for line in open('rentDataset_1D_LR.csv'):
    x, y = line.split(',')
    X.append(x)
    Y.append(y)
```

```python
X = np.array(list(map(int, X)))
Y = np.array(list(map(int, Y)))
```

```python
plt.scatter(X,Y)
plt.xlabel("size (ft^2)")
plt.ylabel("Rental-Cost ($)")
plt.grid(True)
plt.show()
```

```python
m,q = LinearRegression(X,Y)
```

```python
PlotGraph(X,Y,m,q)
```
